1. What is the difference between a function and a method in Python?
   - Function -> A function is a block of reusable code that performs a
     specific task. It is defined using the def keyword and can be called independently. It does not necessarily belong to a class or object.
   - Method -> A method is a function that is associated with an object
     and operates on that object. It is defined inside a class and takes self (or cls for class methods) as its first parameter. It can only be called on an instance (object) of that class.

```python
#function
def greet(name):
    return f"Hello, {name}!"

print(greet("Soul"))

#method
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, {self.name}!"

p = Person("Soul")
print(p.greet())  # Output: Hello, Soul!
```

3. What are the different ways to define and call a function in Python?
   - In Python, functions can be defined and called in multiple ways.

     1. Positional Arguments -> Arguments are passed in order.
     2. Keyword Arguments -> Arguments are passed with parameter names.
     3. Default Arguments -> Parameters have default values if not
        provided.
     4. Variable-Length Arguments (*args) -> Used to pass multiple arguments as a tuple.
     5. Keyword Variable-Length Arguments (**kwargs) -> Used to pass multiple keyword arguments as a dictionary.
     6. Lambda Functions (Anonymous Functions) -> Single-line functions using lambda.
     7. Function with Return Statement -> Returns a value to the caller.

```python
#positional arguments
def add(a, b):
    return a + b

result = add(3, 5)


#Keyword arguments
def greet(name, msg):
    print(f"{msg}, {name}!")

greet(name="Soul", msg="Hello")



#Default arguments
def greet(name, msg="Hi"):
    print(f"{msg}, {name}!")

greet("Soul")



#variable-length arguments
def sum_all(*args):
    return sum(args)

print(sum_all(1, 2, 3, 4))



#Keyword variable-length arguments
def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_details(name="Soul", age=25)



#Lambda functions
square = lambda x: x ** 2
print(square(4))



#Functions with Return statement
def multiply(a, b):
    return a * b

result = multiply(3, 4)

```

4. What is the purpose of the 'return' statement in a Python function?
    - The return statement in a Python function serves the following purposes:
      - Returns a Value: It specifies the value that the function should
        return to the caller. Without a return statement, a function returns None by default.
      - Ends Function Execution: Once a return statement is encountered,
        the function immediately stops executing further code.
      - Allows Reusability: By returning values, functions can be used
        dynamically in expressions, assignments, and other function calls.

```python
def square(num):
    return num ** 2

result = square(5)
print(result)

#early termination
def check_even(num):
    if num % 2 == 0:
        return "Even"
    return "Odd"

print(check_even(4))
print(check_even(7))

```

5. What are iterators in Python and how do they differ from iterables?
   -  Iterable -> An iterable is an object that contains multiple
        elements and can be looped over (e.g., lists, tuples, strings, dictionaries, sets).It has a special method __iter__() that returns an iterator.
   -  Iterator -> An iterator is an object that remembers its position
        and produces the next value when asked.It must have two special methods:
       - __iter__() → Returns the iterator itself.
       - __next__() → Returns the next element in the sequence,     raising StopIteration when no more elements are left.

```python
#iterable

numbers = [1, 2, 3, 4]  # A list is an iterable
for num in numbers:
    print(num)

#iterators

numbers = iter([1, 2, 3])  # Convert list to an iterator
print(next(numbers))
print(next(numbers))
print(next(numbers))
# print(next(numbers))  # Raises StopIteration

```

6. Explain the concept of generators in Python and how they are defined.
   - A generator is a special type of iterator that simplifies the process
     of creating iterators using functions instead of classes. They are defined using the yield keyword.
      - When a function contains yield, it becomes a generator function.

      - Calling this function does not execute it immediately; instead, it
        returns a generator object.

      - Each time next() is called, execution resumes from where it left   off, yielding the next value.

      - Once all values are generated, it raises a StopIteration exception.

  -  Advantages of Generators
      - ✔ Memory Efficient: Ideal for large datasets (e.g., reading large files line by line).
      - ✔ Faster Execution: Generates values lazily instead of storing all at once.
      - ✔ State Retention: Keeps track of where it left off.

```python
def square_numbers(n):
    for i in range(n):
        yield i ** 2  # Instead of returning, it yields values one by one

gen = square_numbers(3)  # Creates a generator object

print(next(gen))
print(next(gen))
print(next(gen))
# print(next(gen))  # Raises StopIteration
```

7. What are the advantages of using generators over regular function?
   - Advantages of Generators Over Regular Functions
      - Memory Efficient – Generates values on demand, avoiding storing large datasets in memory.
      - Faster Execution – No need to create and store lists; values are produced lazily.
      - State Retention – Remembers where it left off, allowing efficient iteration.
      - Simpler Code – No need for custom iterator classes; yield makes implementation easier.
      - Streaming Support – Ideal for processing large files or infinite sequences.

```python
#regular function
def generate_numbers(n):
    numbers = []  # Stores all numbers in memory
    for i in range(n):
        numbers.append(i)
    return numbers  # Returns the entire list

nums = generate_numbers(10**6)  # Stores 1 million numbers in memory
print(nums[:5])

#generators
def generate_numbers(n):
    for i in range(n):
        yield i  # Generates values on demand

gen = generate_numbers(10**6)  # No large memory usage
print(next(gen))
print(next(gen))
print(next(gen))
```

8. What is a Lambda function in Python and when is it typically used?
   - A lambda function is a small, anonymous function defined using the
     lambda keyword. It can take multiple arguments but only contain a single expression.
   - Lambda functions are typically used when:
       - You need a small function for a short period (without defining a full function using def).
       - You need a quick function inside higher-order functions like map(), filter(), or sorted().
       - You want to keep the code concise (useful in functional programming).
   - Key Advantages of Lambda Functions
      - ✔ Concise & Readable – Reduces boilerplate code.
      - ✔ Useful for one-time use functions – No need to define a full function.
      - ✔ Great for functional programming – Works well with map(), filter(), sorted(), etc.

```python
#basic lambda function
square = lambda x: x ** 2
print(square(5))  # Output: 25

#using lambda in map()
nums = [1, 2, 3, 4]
squared = list(map(lambda x: x ** 2, nums))
print(squared)  # Output: [1, 4, 9, 16]

#using lambda in filter()
nums = [1, 2, 3, 4, 5]
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)  # Output: [2, 4]

#using lambda in sorted()
pairs = [(1, 3), (2, 2), (4, 1)]
sorted_pairs = sorted(pairs, key=lambda x: x[1])
print(sorted_pairs)  # Output: [(4, 1), (2, 2), (1, 3)]
```

9. Explain the purpose and usage of 'map()' function in Python.
   - Purpose of map() in Python is The map() function applies a given function to each item in an iterable (e.g., list, tuple) and returns a map object (iterator).

   - Key Advantages of map()
      -  Concise & Readable – Eliminates the need for explicit loops.
      -  Memory Efficient – Returns an iterator instead of creating a new list.
      -  Works Well with Lambda Functions – Ideal for quick transformations.

```python
#syntax>>>>map(function, iterable)

#squaring a list of numbers
nums = [1, 2, 3, 4]
squared = list(map(lambda x: x ** 2, nums))
print(squared)  # Output: [1, 4, 9, 16]

#converting strings to upper case
words = ["hello", "world"]
upper_words = list(map(str.upper, words))
print(upper_words)  # Output: ['HELLO', 'WORLD']

#adding two list element-wise
a = [1, 2, 3]
b = [4, 5, 6]
sum_list = list(map(lambda x, y: x + y, a, b))
print(sum_list)  # Output: [5, 7, 9]
```

10. What is the difference between 'map()','reduce()' and 'filter()'
     functions in Python?
     - The map(), filter(), and reduce() functions in Python are all used
       for processing iterables but serve different purposes. The map() function applies a given function to each element in an iterable and returns an iterator with the transformed values. It is useful for modifying data, such as squaring numbers in a list. On the other hand, filter() is used to select elements from an iterable based on a condition. It applies a function that returns True or False and only includes elements that satisfy the condition, making it ideal for tasks like extracting even numbers from a list. Lastly, reduce(), which comes from the functools module, is used to combine elements of an iterable into a single value by repeatedly applying a function. It is commonly used for tasks like summing up numbers in a list. While map() and filter() return iterators with multiple elements, reduce() returns a single aggregated result.
         - Use map() when to transform all elements.
         - Use filter() when to select specific elements.
         - Use reduce() when to combine elements into a single
           value.

```python
#map()
nums = [1, 2, 3, 4]
squared = list(map(lambda x: x ** 2, nums))
print(squared)  # Output: [1, 4, 9, 16]

#filter()
nums = [1, 2, 3, 4, 5]
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)  # Output: [2, 4]

#reduce
from functools import reduce
nums = [1, 2, 3, 4, 5]
sum_all = reduce(lambda x, y: x + y, nums)
print(sum_all)  # Output: 15
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example usage:
for num in range(1, 201):
    if is_prime(num):
        print(num, end=" ")
```

5. Create an iterator class in Python that generates the Fibonacci sequence upto a specified number of terms.

```python
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms  # Number of terms
        self.a, self.b = 0, 1   # First two Fibonacci numbers
        self.count = 0          # Counter to track iterations

    def __iter__(self):
        return self  # Returns the iterator object

    def __next__(self):
        if self.count >= self.n_terms:  # Stop when reaching the limit
            raise StopIteration
        fib = self.a
        self.a, self.b = self.b, self.a + self.b  # Update Fibonacci values
        self.count += 1
        return fib

# Example usage:
fib_seq = FibonacciIterator(10)  # Generate first 10 Fibonacci numbers
for num in fib_seq:
    print(num, end=" ")
```

6. Write a generator function in Python that yields the power of 2 upto a given exponent.

```python
def power_of_two(exponent):
    for i in range(exponent + 1):  # Loop from 0 to exponent
        yield 2 ** i  # Yield power of 2

# Example usage:
for num in power_of_two(5):
    print(num, end=" ")
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_path):
    with open(file_path, "r") as file:  # Open file in read mode
        for line in file:
            yield line.strip()  # Yield each line without trailing newline

# Example usage:
file_path = "sample.txt"  # Replace with the file path
for line in read_file_line_by_line(file_path):
    print(line)  # Prints each line one by one
```

9. Write a Python program that uses 'map()' to convert a list of temperatures from Celsius to Fahrenheit.

```python
# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temps = [0, 10, 20, 30, 40, 100]

# Convert using map()
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

# Output the converted temperatures
print(fahrenheit_temps)
```

10. Create a Python program that uses 'filter()' to remove all the vowels from a given string.

```python
def remove_vowels(s):
    vowels = "aeiouAEIOU"
    return "".join(filter(lambda char: char not in vowels, s))

# Example usage
input_string = "Hello, World!"
result = remove_vowels(input_string)
print(result)
```
